```python
def temperature_modeling(a, b, c, time):
    temperature = a * time**2 + b * time + c
    return temperature

a_hardcoded = 0.1
b_hardcoded = 2
c_hardcoded = 10

time_hardcoded = 5
print(temperature_modeling(a_hardcoded, b_hardcoded, c_hardcoded, time_hardcoded))

a_keyboard = float(input("Enter coefficient a: "))
b_keyboard = float(input("Enter coefficient b: "))
c_keyboard = float(input("Enter coefficient c: "))
time_keyboard = float(input("Enter time in hours: "))

print(temperature_modeling(a_keyboard, b_keyboard, c_keyboard, time_keyboard))

with open("weather_data.txt", "w") as file:
    file.write("0.2 1.5 8\n")
    file.write("4")

with open("weather_data.txt", "r") as file:
    lines = file.readlines()
    a_file, b_file, c_file = map(float, lines[0].split())
    time_file = float(lines[1])

print(temperature_modeling(a_file, b_file, c_file, time_file))
```
